```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```python
df = pd.read_csv("MLdata_Schaeffer.csv").set_index('case')
df
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(df.corr(), cmap="magma", ax=ax)
plt.savefig("Graphics\CorrPlot_Schaeffer.png")
```

```python
df.corr()['outcome'].to_frame().sort_values(by='outcome', ascending=False)
```

```python
sns.histplot(df['mean'])
```

```python
sns.jointplot(x='mean', y='median', data=df, height=8, color="blue")
plt.title("Mean vs. Median Pixel Value in 900-1000 Xray Range")
plt.savefig("Graphics\MeanVsMedianPixelValues.jpg")
```
