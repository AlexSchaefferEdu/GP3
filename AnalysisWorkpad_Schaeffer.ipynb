```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```python
Schaeffer = pd.read_csv("MLdata_Schaeffer.csv").drop(['case', 'prediction'], axis=1)
fig, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(Schaeffer.corr(), cmap="magma", ax=ax)
plt.title("Schaeffer Sample")
plt.savefig("Graphics\CorrPlot_Schaeffer.png")
```

```python
Stack = pd.read_csv("MLdata_Stack.csv").drop('case', axis=1)
fig, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(Stack.corr(), cmap="magma", ax=ax)
plt.title("Stack Sample")
plt.savefig("Graphics\CorrPlot_Stack.png")
```

```python
Vuaghn = pd.read_csv("MLdata_Vuaghn.csv").drop('case', axis=1)
fig, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(Vuaghn.corr(), cmap="magma", ax=ax)
plt.title("Vuaghn Sample")
plt.savefig("Graphics\CorrPlot_Vuaghn.png")
```

```python
Terry = pd.read_csv("MLdata_Terry.csv").drop('case', axis=1)
fig, ax = plt.subplots(figsize=(8, 8))
sns.heatmap(Terry.corr(), cmap="magma", ax=ax)
plt.title("Terry Sample")
plt.savefig("Graphics\CorrPlot_Terry.png")
```

```python
Part = pd.concat([Schaeffer, Stack, Vuaghn, Terry], axis=0).reset_index().drop('index', axis=1)
Part.describe()
```
